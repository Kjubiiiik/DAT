# Analýza sloupce
- **Data**: Sloupec obsahuje technické specifikace a doplňkové parametry produktů, jako jsou značky, typy procesorů, kapacity RAM, barvy, rozlišení nebo rozměry.
- **Struktura**: Jedná se o formát **JSON** (objekt klíč-hodnota). Každý záznam obsahuje klíč *"category"*, ale ostatní klíče jsou dynamické a mění se podle typu zboží (např. u monitorů *"size_inch"*, u kabelů *"length_m"*).

# 2.
# Analýza dotazu s JSON_VALUE

```python
SELECT
    name,
    JSON_VALUE(description, '$.category') AS category
FROM product;
```

### Popis funkce:
- Tento dotaz vybere název produktu a k němu vypíše konkrétní hodnotu, která je uložena jako *"category"*.

# 3.
# SQL Skripty 
### 1. Zobrazení produktů z vybraného skladu


```python
SELECT p.name, p.quantity, w.name AS warehouse_name
FROM product p
JOIN warehouse w ON p.warehouse_id = w.id
WHERE w.name = 'Central Warehouse';
```

### 2. Produkty s kategorií *"cable"* v popisu

```python
SELECT name, quantity, description
FROM product
WHERE JSON_VALUE(description, '$.category') = 'cable';
```

### 3. Produkty s množstvím > 20 (vzestupně)

```python
SELECT name, quantity
FROM product
WHERE quantity > 20
ORDER BY quantity ASC;
```

### 4. Produkty uskladněné v březnu 2024

```python
SELECT name, quantity, stored_from
FROM product
WHERE stored_from BETWEEN '2024-03-01' AND '2024-03-31';
```
